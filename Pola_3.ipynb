```python
#Scatter Plot

data1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Praktikum/tips.csv")

# Scatter plot with day against tip
plt.scatter(data1['sex'], data1['tip'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Sex')
plt.ylabel('Tip')

plt.show()
```
